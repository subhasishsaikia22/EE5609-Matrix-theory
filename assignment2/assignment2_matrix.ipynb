```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#x1 are the x coordinates of the points for the line between A and B, y1 are the y coordinates for the same vertices.
#x2,x3,x4 and y2,y3,y4 are the same for the other line
x1, y1 = [2, 5], [1, 4]
x2, y2 = [5, 4], [4, 7]
x3, y3 = [4, 1], [7, 4]
x4, y4 = [1, 2], [4, 1]
plt.plot(x1, y1, x2, y2,x3,y3,x4,y4, marker = 'o')
A=np.array(([2,1]))
B=np.array(([5,4]))
C=np.array(([4,7]))
D=np.array(([1,4]))
#labelling the coordinates
Para_coords = np.vstack((A,B,C,D)).T
vert_labels = [A,B,C,D]
for i, txt in enumerate(vert_labels):
    plt.annotate(txt, # this is the text
                 (Para_coords[0,i], Para_coords[1,i]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,3), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid()
plt.show()
```
