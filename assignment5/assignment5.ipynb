```python
import numpy as np
import math
import matplotlib.pyplot as plt
from numpy import linalg as LA

#Input parameters
x = np.linspace(-6,12,100)         
q = np.array([3,4])
O = np.array([0,0])
f =-25
r= np.sqrt(LA.norm(O)**2-f)

# equation of line 
# y = (-a/b)*x + c/b
y1 = (-3/4)*x + 25/4
y2=(4/3)*x
x4 = [q[0],O[0]]
y4 = [q[1],O[1]]

#Generating points on a circle
def circ_gen(O,r):
	len = 50
	theta = np.linspace(0,2*np.pi,len)
	x_circ = np.zeros((2,len))
	x_circ[0,:] = r*np.cos(theta)
	x_circ[1,:] = r*np.sin(theta)
	x_circ = (x_circ.T + O).T
	return x_circ
##Generating the circle
x_circ1= circ_gen(O,r)
#Plotting the line
plt.plot(x, y1,label='Eqn of the tangent:3x+4y=25')
plt.plot(x, y2,label='Eqn of the normal:4x-3y=0')
plt.plot(x4,y4,'og')
#Plotting the circle
plt.plot(x_circ1[0,:],x_circ1[1,:],label='$given curve: x^2+y^2=25$')

#labelling the coordinates
Para_coords = np.vstack((q,O)).T
vert_labels = [q,O]
for i, txt in enumerate(vert_labels):
    plt.annotate(txt, # this is the text
                 (Para_coords[0,i], Para_coords[1,i]), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,3), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
plt.xlabel('$x-axis$')
plt.ylabel('$y-axis$')
plt.xlim(-10, 20)
plt.ylim(-10, 20)
plt.axis('equal')
plt.grid()
plt.legend(loc='best')
plt.show()
```
